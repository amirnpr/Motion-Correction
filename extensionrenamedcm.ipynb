```python
import os

def rename_no_ext_to_dcm(directory):
    # Walk through the directory and its subdirectories
    for foldername, subfolders, filenames in os.walk(directory):
        for filename in filenames:
            # Check if the file has no extension
            if '.' not in filename:
                no_ext_path = os.path.join(foldername, filename)
                # Create new file name with .dcm extension
                dcm_path = os.path.join(foldername, filename + '.dcm')
                # Rename the file
                os.rename(no_ext_path, dcm_path)
                print(f'Renamed: {no_ext_path} -> {dcm_path}')

# Provide the path to the folder '3'
directory_path = r"V:\dataset\F1"

# Call the function
rename_no_ext_to_dcm(directory_path)

```
